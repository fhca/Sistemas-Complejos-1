```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

R = 2.5
x = np.arange(-10, 10, .01)

def f(x):
    return R * np.sin(x)

plt.plot(x, f(x), x, x)


def orbita(x0, f, nveces):
    orb = np.zeros(nveces + 1)
    orb[0] = x0
    for i in range(nveces):
        orb[i+1] = f(orb[i])
    return orb

def telaraÃ±a(x0, f, nveces):
    orb = orbita(x0, f, nveces)
    x = np.repeat(orb, 2)
    y = np.roll(x, -1) # desplaza elem de x un lugar a la izq
    y[0] = 0
    y[-1] = f(x[-1])
    plt.plot(x, y, alpha=0.5)

telaraÃ±a(1, f, 1000)
telaraÃ±a(-1, f, 1000)
telaraÃ±a(-6, f, 1000)
```
