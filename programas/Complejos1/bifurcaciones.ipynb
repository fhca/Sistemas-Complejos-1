```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

#from matplotlib import cm

def logistic(c, x):
    return c * x * (1 - x)

def seno(c, x):
    return c * np.sin(x)

funcin = logistic
plt.figure(figsize=(30,30))

N=2000  # N x N square window, (resolution)
C = np.linspace(3.5, 4, N)
Y = np.array([0.5] * N)
#Y[0::2]*=-1  para seno
for _ in range(500):
    Y = funcin(C, Y)  # wasted initial values
for _ in range(N):
    Y = funcin(C, Y)
    plt.plot(C, Y, 'k,')
plt.show()
```

```python
# dimensin puntual

N=2000  # N x N square window, (resolution)
C = np.linspace(3.5, 4, N)
Y = np.array([0.5] * N)

rx = np.linspace(3.6,4,200)  # center coords
ry = np.linspace(.4,.9,200)
Nx = np.zeros(200)  # of points close to centers

for _ in range(500):
    Y = funcin(C, Y)  # wasted initial values
for _ in range(N):
    Y = funcin(C, Y)
    count = 0
    for e in np.linspace(.2, .8, 200):
        for x,y in np.stack((rx, ry), axis=1):
            Nx[count] += np.sum(np.sqrt((C-x)**2+(Y-y)**2) < e)
        count+=1
plt.plot(np.linspace(.2, .8, 200), Nx)
plt.show()
```
