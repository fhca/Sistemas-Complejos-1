```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

#from matplotlib import cm

def logistic(c, x):
    return c * x * (1 - x)

def seno(c, x):
    return c * np.sin(x)

funcin = logistic
plt.figure(figsize=(20,20))

N=2000  # N x N square window, (resolution)
C = np.linspace(3.5, 4, N)
Y = np.array([-0.5] * N)
Y[0::2]*=-1
for _ in range(500):
    Y = funcin(C, Y)  # wasted initial values
for _ in range(N):
    Y = funcin(C, Y)
    plt.plot(C, Y, 'k,', alpha=0.5)
plt.show()
```
