```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

#notebook
```

```python
np.arange(0.01, 10*np.pi, .01)
```

```python
t = np.arange(0.01, 10*np.pi, .01)
x = t*np.sin(t)

plt.plot(x)
```

```python
tau = 350
plt.plot(x[:-tau], x[tau:])
plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
tau = 50
ax.plot(x[:-2*tau], x[tau:-tau], zs=x[2*tau:])
```

```python
t = np.arange(0.01, 10*np.pi, .01)
x = np.sin(t)/t

plt.plot(x)
```

```python
tau = 150
plt.plot(x[:-tau], x[tau:])
plt.show()
```

```python
%matplotlib notebook
plt.rcParams["animation.html"] = "html5"

t = np.arange(.01, 10*np.pi, .01)
x = np.sin(t)+2*np.sin(2*t)+np.sin(3*t)
#plt.plot(x)
#plt.show()

fig, ax = plt.subplots()
ax.axis([-4,4,-4,4])
l, = ax.plot([],[])

tau = list(range(10, 640, 10))

def anim(i):
    l.set_data(x[:-tau[i]], x[tau[i]:])

from matplotlib import animation
ani = animation.FuncAnimation(fig, anim, frames=len(tau))

```

```python
%matplotlib notebook
plt.rcParams["animation.html"] = "html5"
from mpl_toolkits.mplot3d import Axes3D

t = np.arange(.01, 10*np.pi, .1)
x = np.sin(t)+2*np.sin(2*t)+np.sin(3*t)
#plt.plot(x)
#plt.show()

fig, ax = plt.subplots()
ax = Axes3D(fig)
ax.axis([-4,4,-4,4])
ax.set_zlim(-4,4)

tau = list(range(1, 64, 1))

l, = ax.plot([],[],[])
tau_text = ax.text(0,0,3.3, '', transform=ax.transAxes)
tau_text.set_text('')

def anim(i): # parametrizar con índice de tau
    t=tau[i]
    l.set_data(x[:-2*t], x[t:-t])
    l.set_3d_properties(x[2*t:])
    tau_text.set_text("tau = {}".format(t))
    return fig,

from matplotlib import animation
ani=animation.FuncAnimation(fig, anim, frames=len(tau), blit=True)


```
